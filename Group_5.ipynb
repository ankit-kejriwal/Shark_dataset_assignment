```python
# final_data["new_Activity"] = final_data['Activity'].str.contains('swimming', regex=False)

# activity_series = final_data['Activity']
# activity_series
# activity_series[activity_series.str.contains('swimming', regex=False)]

import re
import numpy as np
# final_data.loc[final_data.Activity.str.contains('fishing',case=False, regex=True), 'Activity'] = 'xyz'

fishing = 'fishing'
surfing = 'surfing'
swimming = 'swimming'
boarding = 'boarding'
others = 'others'
def replace_activity(x):
    
    if fishing in x:
        return "Fishing"
    elif surfing in x:
        return "Surfing"
    elif swimming in x:
        return "Swimming"
    elif boarding in x:
        return "Boarding"
    else:
        return "Others"
```

```python
final_data['Activity_NewCol'] = final_data.apply(lambda x: replace_activity(x['Activity'].lower()), axis=1)
```

```python
final_data.groupby('Activity_NewCol').size().sort_values(ascending=False)
```
